```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
import matplotlib.pyplot as plt
import warnings
import seaborn as sns

url = "https://raw.githubusercontent.com/paulopadilla/digital_mkt/main/taller3/boletas_20000.csv"
data = pd.read_csv(url, sep=';')


url_margen = "https://raw.githubusercontent.com/paulopadilla/digital_mkt/main/taller3/margen_categorias.csv"



# eliminar datos duplicados
num_duplicadas = data.duplicated().sum()
data.drop_duplicates(inplace=True)

# Verificar y contar valores nulos
nulos = data.isnull().sum().sum()

# Imprimir resultados
print(f"\n Filas duplicadas eliminadas: {num_duplicadas if num_duplicadas > 0 else '0'}")
if data.isnull().any().any():
  print("\n Hay datos nulos en la BBDD, aquí el desglose: \n", data.isnull().sum())
else:
  print ("\n No hay datos nulos en la BBDD\n")

print("Info de los datos:")
print(data.info())
print(data.head())
```

# Pregunta 1
#### Corra el algoritimo apriori sobre la BD transaccional usando 1000, 100 y 10 transacciones como soporte mínimo. Reporte el tiempo de procesamiento y número de reglas extraídas. Explique el resultado.

```python
#1000 transacciones mínimas
soporte_min_1000 = 1000/data.shape[0]
frequent_itemsets_1000 = apriori(data.iloc[:, 2:], min_support = soporte_min_1000, use_colnames = True, max_len = 2)
display(frequent_itemsets_1000)
```

```python
rules_1000 = association_rules(frequent_itemsets_1000, metric="lift", min_threshold = 1)
```

```python
#100 transacciones mínimas
soporte_min_100 = 100/data.shape[0]
frequent_itemsets_100 = apriori(data.iloc[:, 2:], min_support = soporte_min_100, use_colnames = True, max_len = 2)
display(frequent_itemsets_100)
```

```python
rules_100 = association_rules(frequent_itemsets_100, metric="lift", min_threshold = 1)
```

```python
#10 transacciones mínimas
soporte_min_10 = 10/data.shape[0]
frequent_itemsets_10 = apriori(data.iloc[:, 2:], min_support = soporte_min_10, use_colnames = True, max_len = 2)
display(frequent_itemsets_10)
```

```python
rules_10 = association_rules(frequent_itemsets_10, metric="lift", min_threshold = 1)
```

Se puede observar que con la condición de 1000 transacciones mínimas se encontraron 149 reglas y el tiempo de ejecución fue de 1 segundo. Para 100 transacciones minimas se obtuvieron 3080 reglas y el tiempo de ejecución fue de 2 segundos. Por último, para 10 transacciones se encontraron 12659 reglas y se demoró 21 segundos. Este cambio se debe a que, mientras la condición baje, se van a encontrar más transacciones que cumplan, por lo que la tarea se demorará más en terminar.

# Pregunta 2
#### Vuelva a correr a priori usando 20 transacciones como soporte mínimo y reporte las 20 reglas entre dos categorías, con mayor soporte, confianza y ganancia. Interprete y explique los valores para la regla con mayor soporte, confianza y ganancia.

```python
soporte_min_20 = 20/data.shape[0]
frequent_itemsets_20 = apriori(data.iloc[:, 2:], min_support = soporte_min_20, use_colnames = True, max_len = 2)
display(frequent_itemsets_20)
```

```python
rules_20 = association_rules(frequent_itemsets_20, metric="lift", min_threshold = 1)
print("Número de reglas: ", rules_20.shape[0])

rules_20.head(5)
```

```python
rules_20_soporte = rules_20.sort_values('support', ascending= False)
rules_20_soporte.head(20)
```

Al ordenar según el mayor soporte, el primer lugar corresponde a yoghurt y galletas dulces. Esto significa que, dentro de las compras analizadas, un 0.133507 (o un 13%) contiene yoghurt y galletas dulces.

```python
rules_20_confianza = rules_20.sort_values('confidence', ascending= False)
rules_20_confianza.head(20)
```

Para la regla con mayor confianza se encuentra comino y salsa de tomate con un valor de 0.793103. Por lo tanto, dentro de las compras que llevan comino, un 79% también lleva salsa de tomate. Es importante destacar que la salsa de tomate aparece 6 veces como consecuente dentro de las 20 reglas, por lo que este resultado indica que es un producto común. Probablemente no hay una gran cantidad de compras que incluyan comino, resultando en que muchas de estas compras coincidentalmente también incluyen salsa de tomate.

```python
rules_20_ganancia = rules_20.sort_values('lift', ascending= False)
rules_20_ganancia.head(20)
```

En el caso de la mayor ganancia, implementos de cumpleaños y varios utencilios de cocina obtienen el mayor valor, con 27.308603. Esto significa que cuando alguien lleva implementos de cumpleaños, es 27.3 veces más probable que también lleve utencilios de cocina, y viceversa.
