```python

import pandas as pd

url = "https://raw.githubusercontent.com/paulopadilla/digital_mkt/main/taller3/boletas_20000.csv"
data = pd.read_csv('boletas_20000.csv')


# eliminar datos duplicados 
num_duplicadas = data.duplicated().sum()
data.drop_duplicates(inplace=True)
# Verificar y contar valores nulos
nulos = data.isnull().sum().sum()

# Imprimir resultados
print(f"\n Filas duplicadas eliminadas: {num_duplicadas if num_duplicadas > 0 else '0'}")
if data.isnull().any().any():
  print("\n Hay datos nulos en la BBDD, aqu√≠ el desglose: \n", data.isnull().sum())
else:
  print ("\n No hay datos nulos en la BBDD\n")

print("Info de los datos:")
print(data.info())
print(data.head())
```
